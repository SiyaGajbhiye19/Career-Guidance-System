```python
from sklearn.preprocessing import LabelEncoder

# Make a copy of the original DataFrame to store the encoded values
data_encoded = df_cleaned.copy()

# Label encode the 'Role' column
le = LabelEncoder()
data_encoded['Role'] = le.fit_transform(data_encoded['Role'])

# Define the mapping for ordinal encoding for skill levels
category_mapping = {
    "Not Interested": 0,
    "Poor": 1,
    "Beginner": 2,
    "Average": 3,
    "Intermediate": 4,
    "Excellent": 5,
    "Professional": 6
}

# Select all columns except 'Role'
columns_to_encode = data_encoded.columns[:-1]  # Exclude the 'Role' column

# Apply the map function to each skill-related column
for column in columns_to_encode:
    data_encoded[column] = data_encoded[column].map(category_mapping)

# Display the first few rows of the encoded data
print(data_encoded.head())
```
