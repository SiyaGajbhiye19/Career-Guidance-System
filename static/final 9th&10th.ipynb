```python
# Function to recommend a stream based on all features
def recommend_stream(row):
    # Grades mapping: Convert to numerical scores for comparison
    grade_mapping = {'A': 5, 'B': 4, 'C': 3, 'D': 2, 'E': 1}
    math_grade = grade_mapping[row['Grades in Mathematics']]
    science_grade = grade_mapping[row['Grades in Science']]
    english_grade = grade_mapping[row['Grades in English']]
    social_grade = grade_mapping[row['Grades in Social studies']]
    cs_grade = grade_mapping[row['Grades in Computer Science/Information Technology']]

    # Interests mapping: Assign weights to interest levels
    interest_mapping = {'No Interest': 0, 'Low Interest': 1, 'Moderate Interest': 2, 'High Interest': 3}
    math_interest = interest_mapping[row['Interest in Mathematics']]
    science_interest = interest_mapping[row['Interest in Science']]
    english_interest = interest_mapping[row['Interest in English']]
    social_interest = interest_mapping[row['Interest in Social Studies']]
    cs_interest = interest_mapping[row['Interest In Computer Science/Information Technology']]
    new_concept_interest = interest_mapping[row['Interest in Learning New Concepts/Technology']]

    # Additional features
    activities = row['Extracurricular Activities']
    career_preference = row['Preferred Career Options']
    stream_preference = row['Preferred Stream for 11th Grade']
    soft_skills = row['Soft Skills']

    # Logic for stream recommendation
    # Science (Mathematics)
    if math_grade >= 4 and science_grade >= 4 and math_interest >= 2 and 'Coding' in activities:
        return 'Science (Mathematics)'
    
    # Science (Biology)
    if science_grade >= 4 and science_interest >= 2 and math_grade < 4:
        return 'Science (Biology)'
    
    # Commerce
    if 'Business' in career_preference or 'Leadership' in soft_skills:
        return 'Commerce'
    
    # Arts
    if english_grade >= 4 and english_interest >= 2 or 'Art' in activities or 'Music' in activities:
        return 'Arts'

    # Default to Undecided
    return 'Undecided'

# Load the dataset
# file_path = 'path_to_your_file.xlsx'  # Replace with your file path
# df = pd.read_excel(file_path)

# Apply the function to recommend streams
df['Recommended Stream'] = df.apply(recommend_stream, axis=1)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize base estimator
base_estimator = DecisionTreeClassifier()

# Define Bagging Classifier (use 'estimator' instead of 'base_estimator')
bagging_clf = BaggingClassifier(estimator=base_estimator, random_state=42)
```

```python
bagging_clf.score(X_train,Y_train)*100
```

```python
y_pred = bagging_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
import pandas as pd

# Assuming y_test and y_pred are already defined
# y_pred is the predicted values for the test set, y_test is the actual values

print("For BaggingClassifier")
print("-" * 50)

# Get the number of unique classes in the target variable
n_classes = len(set(y_test))

# Confusion Matrix
cf = confusion_matrix(y_test, y_pred)

# Convert the confusion matrix to a DataFrame
cf_df = pd.DataFrame(cf, columns=[f"Class {i}" for i in range(n_classes)], index=[f"Class {i}" for i in range(n_classes)])

# Print the confusion matrix
print(cf_df)
print("-" * 50)

# Test dataset evaluation metrics
print(f"Test dataset accuracy_score: {accuracy_score(y_test, y_pred)*100:.4f}")
print(f"Test dataset f1_score: {f1_score(y_test, y_pred, average='weighted')*100:.4f}")
print(f"Test dataset recall_score: {recall_score(y_test, y_pred, average='weighted')*100:.4f}")
print(f"Test dataset precision_score: {precision_score(y_test, y_pred, average='weighted')*100:.4f}")
```
